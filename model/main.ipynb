```python
# regression_data.plot.scatter(x='close', y='future_return')
```

```python
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
x = regression_data.loc[:, 'open':]
y = regression_data[['future_return', 'future_risk']]
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2, random_state = 4)
```

```python
regr = linear_model.LinearRegression()
regr.fit(train_x, train_y)

train_y_ = regr.predict(train_x)
print(f'Train R2: {(100 * r2_score(train_y, train_y_)):.2f}%')
test_y_ = regr.predict(test_x)
print(f'Test R2: {(100 * r2_score(test_y, test_y_)):.2f}%')
```

```python
regr = RandomForestRegressor(n_estimators=100)
regr.fit(train_x, train_y)

train_y_ = regr.predict(train_x)
print(f'Train R2: {(100 * r2_score(train_y, train_y_)):.2f}%')
test_y_ = regr.predict(test_x)
print(f'Test R2: {(100 * r2_score(test_y, test_y_)):.2f}%')
```

```python
close_data = selected_data[["date", "symbol", "close"]]
close_data = close_data.pivot(index="date", columns="symbol")
close_data = close_data["close"]
close_data.columns.name = None
close_data
```

```python
# Mean Variance

from pypfopt.expected_returns import mean_historical_return
from pypfopt.risk_models import CovarianceShrinkage
from pypfopt.efficient_frontier import EfficientFrontier

m = mean_historical_return(close_data)  # mu = expected_returns.mean_historical_return(df)
s = CovarianceShrinkage(close_data).ledoit_wolf()  # S = risk_models.sample_cov(df)
ef = EfficientFrontier(m, s)
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()

ef.portfolio_performance(verbose=True)
print(dict(cleaned_weights))
```

```python
# Hierarchical Risk Parity

from pypfopt import HRPOpt

returns = close_data.pct_change().dropna()
hrp = HRPOpt(returns)
weights = hrp.optimize()
cleaned_weights = hrp.clean_weights()

hrp.portfolio_performance(verbose=True)
print(dict(cleaned_weights))
```

```python
# mCVAR

from pypfopt.efficient_frontier import EfficientCVaR

m = mean_historical_return(close_data)
s = close_data.cov()
ef_cvar = EfficientCVaR(m, s)
cvar_weights = ef_cvar.min_cvar()
cleaned_weights = ef_cvar.clean_weights()

ef_cvar.portfolio_performance(verbose=True)
print(dict(cleaned_weights))
```
